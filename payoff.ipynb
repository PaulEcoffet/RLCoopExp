```python
from PartnerChoiceEnv import payoff
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set()
sns.set_palette("colorblind")
sns.set_style("whitegrid")
sns.set_context("paper")
```

```python
x = np.linspace(0, 15, 100)
xESS = 5
xSO = 5 + 5
fig = plt.figure()
ax = fig.gca()
ax.plot(x, payoff(x, xESS))
ax.plot(x, payoff(x, xSO))
ax.plot(x, payoff(x, x))
ax.set_xlim(0, 15)
#ax.axhline(0, c=(0.3, 0.3, 0.3), lw=1)
sns.despine(ax=ax)
plt.show()
```

```python
max_pso = np.max(payoff(x, x))
imso = np.argmax(payoff(x, x))
imess = np.argmax(payoff(x, xESS))
max_pess = np.max(payoff(x, xESS))

print('imax SO:', x[imso], ', max:', max_pso)
print('imax SO:', x[imess], ', max:', max_pess)


```
