```python
from PartnerChoiceEnv import payoff
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set()

a = 5
b = 5
```

```python
x = np.linspace(0, 15, 200)
xESS = a
xSO = a + b
plt.plot(x, payoff(x, xESS))
plt.plot(x, payoff(x, xSO))
plt.plot(x, payoff(x, x))
plt.show()
```

```python
max_pso = np.max(payoff(x, x))
imso = np.argmax(payoff(x, x))
imess = np.argmax(payoff(x, xESS))
max_pess = np.max(payoff(x, xESS))

print('imax SO:', x[imso], ', max:', max_pso)
print('imax SO:', x[imess], ', max:', max_pess)
```
